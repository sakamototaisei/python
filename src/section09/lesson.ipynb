## 問題83. 

3でも7でもない数だったら、その数字を出力するプログラム：

```
a = 4
if a != 3 and a != 7:
    print(f'{a}は3でも7でもありません')
```

上記の条件判定で、andではなくorを使う形に書き換えてください。(※参考 : これをド・モルガンの法則と言います。)

<br>

▼期待する出力
```
4は3でも7でもありません
```

ド・モルガンの法則を使うことで、コードが見やすくなります。
